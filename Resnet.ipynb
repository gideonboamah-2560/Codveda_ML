```python
import tensorflow
base_model = tensorflow.keras.applications.ResNet101(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)


base_model.trainable = False



```

from tensorflow.keras.models import Model
from tensorflow.keras.utils import image_dataset_from_directory


base_model = Model(
    inputs=base_model.input,
    outputs=base_model.layers[-2].output
)

train_dir = r'C:\Users\LAPTOP\Desktop\Tensorflow\brain_tumor_dataset'
train_ds = image_dataset_from_directory(
    train_dir,
    batch_size = 30,
    image_size = (224,224),
    label_mode = 'int'
    
)

num_classes = len(train_ds.class_names)

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python


from tensorflow.keras.layers import Dense

from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, BatchNormalization, Dropout

x = base_model.output
x = GlobalAveragePooling2D()(x)

x = Dense(1024, activation="relu")(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(512, activation="relu")(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

output = Dense(num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=output)
```

```python
import tensorflow

model.compile(
    optimizer = "adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)



```

```python
model.fit(
    train_ds,
    epochs=5
)
```

```python
import numpy


img_path = r'C:\Users\LAPTOP\Desktop\Tensorflow\brain.jfif'
img = tensorflow.keras.utils.load_img(img_path, target_size = (224,224))
img_arr = tensorflow.keras.utils.img_to_array(img)
img_arr = tensorflow.expand_dims(img, axis = 0)

prediction = model.predict(img_arr)
pred = numpy.argmax(prediction[0])
pred_label = train_ds.class_names[pred]
print(pred_label)
```
